# **Data Visualization**

```python
df_global = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv', parse_dates=['Date'])
df_global
```

Print all the countries listed in the data


```python
print(df_global['Country'].unique())
```

```python
countries = ['Argentina', 'Poland', 'United Arab Emirates', 'US', 'Italy', 'Sweden' , 'India']
df_regional = df_global[df_global['Country'].isin(countries)]

```

```python
# Pivot the data for easier plotting
covid = df_regional.pivot(index='Date', columns='Country', values=['Confirmed', 'Recovered', 'Deaths'])
```

## **Now creating the two charts in interactive**

**Below we are prepareing the data for Plotly.**


```python
df_plot = covid['Confirmed'].reset_index().melt(id_vars='Date', var_name='Country', value_name='Confirmed')
df_plot['Recovered'] = covid['Recovered'].reset_index().melt(id_vars='Date', value_name='Recovered')['Recovered']
df_plot['Deaths'] = covid['Deaths'].reset_index().melt(id_vars='Date', value_name='Deaths')['Deaths']
df_plot['Active'] = df_plot['Confirmed'] - df_plot['Recovered'] - df_plot['Deaths']
```



**Now create the line chart**

```python
fig = px.line(df_plot, x='Date', y='Confirmed', color='Country',
              hover_data=['Recovered', 'Deaths', 'Active'])
```

```python
fig.update_layout(
    title='COVID-19 Cases Over Time',
    xaxis_title='Date',
    yaxis_title='Number of Confirmed Cases',
    legend_title='Country',
    hovermode='closest',
    width=1000,
    height=600
)
fig.show()
```

```python
fig.write_html("gdp_canada.html")
```

### **Now creating the Pie chart in interactive way**

```python
# Get the most recent date's data
latest_date = df_regional['Date'].max()
latest_data = df_regional[df_regional['Date'] == latest_date]
```

```python
# Create the pie chart
fig2 = go.Figure(data=[go.Pie(
    labels=latest_data['Country'],
    values=latest_data['Confirmed'],
    hole=.3,
    hoverinfo='label+percent+value',
    textinfo='label+percent'
)])
```

```python
fig2.update_layout(
    title=f'Distribution of Confirmed COVID-19 Cases by Country (as of {latest_date.date()})',
    annotations=[dict(text='Confirmed Cases', x=0.5, y=0.5, font_size=20, showarrow=False)],

    width=1000,
    height=700

)

fig2.show()
```
